**Representa√ß√£o Gr√°fica**

**Total de Vendas por Cidade**

```python
import matplotlib.pyplot as plt

df_grouped_city = df_olist.groupby('customer_city')['price'].sum()


top_cities = df_grouped_city.sort_values(ascending=False).head(10)

# Elabora√ß√£o do gr√°fico de barras
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cidades com o Maior Total de Vendas')
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

**Representa√ß√£o Gr√°fica - Gr√°fico de Barras**

**Total de vendas (R$) por Estados**

```python
import matplotlib.pyplot as plt

df_grouped_state = df_olist.groupby('customer_state')['price'].sum()

#C√≥digo para gr√°fico de barras
plt.figure(figsize=(10, 6))
df_grouped_state.plot(kind='bar', color='lightcoral')
plt.title('Total de Vendas por Estado')
plt.xlabel('Estado')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
