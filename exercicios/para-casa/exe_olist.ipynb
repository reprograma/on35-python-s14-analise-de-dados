```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_pedidos = pd.read_csv('dados/olist_orders_dataset.csv')
df_pedidos
```

```python
df_itens_pedido = pd.read_csv('dados/olist_order_items_dataset.csv')
df_itens_pedido
```

```python
df_olist_pedidos = pd.merge(df_pedidos, df_itens_pedido, on=['order_id'], how='inner')
df_olist_pedidos
```

```python
df_review = pd.read_csv('dados/olist_order_reviews_dataset.csv')
df_review
```

```python
df_olist_pedidos_review = pd.merge(df_olist_pedidos, df_review, on=['order_id'], how='inner')
df_olist_pedidos_review
```

### Limpando os dados

#### Pedidos

```python
df_pedidos.info()
```

```python
columns_drop = ['order_approved_at', 'order_delivered_carrier_date', 'order_estimated_delivery_date']
df_pedidos = df_pedidos.drop(columns_drop, axis=1)
df_pedidos
```

```python
df_pedidos = df_pedidos.loc[df_pedidos['order_status'] == 'delivered']
df_pedidos
```

```python
df_pedidos = df_pedidos.reset_index(drop=True)
df_pedidos
```

```python
df_pedidos.info()
```

```python
df_pedidos['order_purchase_timestamp'] = pd.to_datetime(df_pedidos['order_purchase_timestamp'])
df_pedidos['order_delivered_customer_date'] = pd.to_datetime(df_pedidos['order_delivered_customer_date'])

df_pedidos['reference_month'] = df_pedidos['order_purchase_timestamp'].dt.strftime('%Y-%m-01')
df_pedidos
```

```python
df_pedidos.info()
```

```python
df_pedidos['reference_month'] = pd.to_datetime(df_pedidos['reference_month'])
```

#### itens do pedidos

```python
df_itens_pedido.info()
```

#### reviews

```python
df_review.info()
```

```python
columns_drop = ['review_comment_title', 'review_comment_message']
df_review = df_review.drop(columns_drop, axis=1)
df_review
```

#### Juntando os datasets

```python
df_olist = pd.merge(df_pedidos, df_itens_pedido, on='order_id', how='inner')
df_olist
```

```python
df_olist = pd.merge(df_olist, df_review, on='order_id', how='inner')
df_olist
```

```python
df_olist.to_csv('base_final_excasa_olist.csv', index=False)
```

```python
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de plano
sns.countplot(data = df_olist,
              x = 'review_score')

# definir texto do título e eixos x e y
plt.title('Índice de Satisfação dos Clientes')
plt.xlabel('Índice de Satisafação')
plt.ylabel('Contagem')

plt.show()
```

### 2. O índice de satisfação vem subindo ou diminuindo com o passar do tempo?
Desde o início das atividades da entidade, é possível ver uma variação média mensal das avaliações dos clientes que vai entre 1 e 5. Nota-se que no começo do ano 2018, nos meses de fevereiro e março, a taxa ficou abaixo da média, por volta de 3.7, o que gera certa preocupação, todavia, nos últimos meses, entre junho e agosto, a taxa tem se mantido por volta de 4,2, o que é positivo, mas  precisamos avaliar mais profundamente o que está ocorrendo para que haja toda essa variação no ano de 2018. 

```python
# Agora, vamos fazer uma análise de sazonalidade. Será que o número de vendas aumenta ou diminui de acordo com a época do ano?
df_review_mensal = df_olist.groupby('reference_month').agg({'review_score': 'mean'})
df_review_mensal.sort_values(by='reference_month', ascending=True)
```

```python
# definir tamanho da figura
plt.figure(figsize=(12, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os países
# com tonalidades das barras definidas pela faixa etária

sns.countplot(data = df_olist,
              x = 'reference_month',
              hue = 'review_score')

# definir texto do título e eixos x e y

plt.title('ìndice de Satisfação dos Clientes')
plt.xlabel('Mês')
plt.ylabel('Contagem')

# como o texto do eixo x é longo, podemos rotacioná-lo para melhorar a visualização
plt.xticks(rotation=330)

# plotar e definir texto da legenda
plt.legend(title = 'ìndice de Satisfação')
plt.show()
```
