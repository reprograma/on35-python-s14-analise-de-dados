**3. Crie visualizações que responda as seguintes perguntas:**

**3.1. Qual é o sistema operacional mais usado para cada faixa etária? Existe algum padrão? O tipo de aparelho muda de acordo com o grupo de idade?**

25-29 Tablet

30-34 Smart TV

35-39 Smartphone

40-44 Laptop

45-49 Smart TV

50-54 Laptop

Em geral, há pouca variação no uso dos diferentes dispositivos, todas as faixas etárias usam todos os dispositivos quase que na mesma proporção, tendo uma variação um pouco mais significativa na faixa de 40-44 anos. Embora influencie, a idade não parece ser fator determinante na escolha do dispositivo para este fim. Seria interessante investigar outros fatores de influência.


```python
plt.figure(figsize=(12, 6))

ax = sns.countplot(data = df_netflix,      
              x= 'grupo_idade', 
              hue='Device',
              palette='Set2')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Uso dispositivos por faixas etárias')
plt.xlabel('Faixa Etária')
plt.ylabel('Contagem')
plt.show()
```
