An√°lise dos Dados da Netflix

```python
import pandas as pd
```

```python
netflix_df = pd.read_csv('dados/netflix.csv')
```

```python
print(netflix_df.head())
```

```python
Device_by_Age = netflix_df.groupby(['Age', 'Device']).size().unstack().fillna(0)
print(Device_by_Age)
```

```python
import matplotlib.pyplot as plt
```

```python
netflix_br = netflix_df[netflix_df['Country'] == 'Brazil']
```

```python
Device_count_br = netflix_br['Device'].value_counts()
```

```python
print(Device_count_br)
```

```python
Device_count_br.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Dispositivos mais utilizados no Brasil')
plt.ylabel('')  
plt.show()
```

An√°lise dos Dados da Olist

```python
olist_df = pd.read_csv('dados/olist_products_dataset.csv')
```

```python
print(olist_df.head())
```

```python
missing_values = olist_df.isnull().sum()
print(missing_values)
```

```python
product_category_distribution = olist_df['product_category_name'].value_counts()
```

```python
product_category_distribution.plot(kind='barh', figsize=(10, 10))
plt.title('Distribui√ß√£o de produtos por categoria')
plt.xlabel('Quantidade de Produtos')
plt.ylabel('Categoria')
plt.show()
```
