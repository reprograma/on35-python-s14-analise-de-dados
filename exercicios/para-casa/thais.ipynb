```python
import pandas as pd
```

```python
#Lê o arquivo CSV 'dados/olist_order_items_dataset.csv' e cria um DataFrame chamado 'df_itens'.
df_pedidos = pd.read_csv('dados/olist_order_items_dataset.csv')
df_pedidos
```

```python
#Lê o arquivo CSV 'dados/olist_order_products_dataset.csv' e cria um DataFrame chamado 'df_produtos'.
df_produtos=pd.read_csv('dados/olist_products_dataset.csv')
df_produtos
```

```python
#Lê o arquivo CSV 'dados/olist_order_payments_dataset.csv' e cria um DataFrame chamado 'df_pagamentos'.
df_pagamentos=pd.read_csv('dados/olist_order_payments_dataset.csv')
df_pagamentos
```

```python
#Junta os DataFrames 'df_pedidos' e 'df_produtos' usando a coluna 'product_id' como chave.
# O resultado é armazenado em um novo DataFrame chamado 'df_olist_pedidos'.
df_olist_pedidos=pd.merge(df_pedidos,df_produtos,on='product_id',how='inner')
df_olist_pedidos
```

```python
#Junta os DataFrames 'df_olist_pedidos' e 'df_pagamentos' usando a coluna 'order_id' como chave.
# O resultado é armazenado em um novo DataFrame chamado 'df_olist'.
df_olist=pd.merge(df_olist_pedidos,df_pagamentos, on="order_id",how='inner')
df_olist
```

## ITENS DOS PEDIDOS

```python
#Mostra informações sobre 'df_pedidos'.
#OBS: Todos os dados estão completos e não contêm valores nulos.
df_pedidos.info()
```

Para a tabela order_items, iremos utilizar as seguintes colunas:

```python
#Remove as colunas do DataFrame 'df_pedidos' que não  serão  utilizadas
drop_columns=['order_item_id','seller_id','shipping_limit_date']
df_pedidos=df_pedidos.drop(drop_columns,axis=1)
df_pedidos
```

## PRODUTOS

```python
#Mostra informações sobre 'df_produtos'.
df_produtos.info()
```

Para a tabela olist_products, iremos utilizar as seguintes colunas:

- **product id**
- **product_category_name**



```python
#Remove as colunas do DataFrame 'df_produtos' que não  serão  utilizadas
drop_columns2=['product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm','product_height_cm','product_width_cm']
df_produtos=df_produtos.drop(drop_columns2,axis=1)
df_produtos
```

```python
#Identifica valores nulos
df_produtos.isnull().sum()
```

```python
#Trata valores nulos
df_produtos_itens=df_produtos.fillna('None')
df_produtos_itens.isnull().sum()
```

## PAGAMENTOS

```python
#Mostra informações sobre 'df_pagamentos'.
df_pagamentos.info()
```

Para a tabela olist_payments, iremos utilizar as seguintes colunas:

- **order_id**
- **payment_type**
- **payment_sequential**


```python
#Remove as colunas do DataFrame 'df_pagamentos' que não  serão  utilizadas
drop_columns3=['payment_installments','payment_value']
df_pagamentos=df_pagamentos.drop(drop_columns3,axis=1)
df_pagamentos
```

## Juntando os três datasets

```python
df_olist = pd.merge(df_pedidos, df_produtos, on='product_id', how='inner')
df_olist = pd.merge(df_olist, df_pagamentos, on='order_id', how='inner')
df_olist
```

## Analisando os dados

- Qual a quantidade de brinquedos comprados no valor abaixo de 100,00 e acima de 100,00, respectivamente?

**2817 brinquedos abaixo de R$ 100,00**

**1462 brinquedos acima de R$ 100,00**

- Qual a porcentagem do frete em relação ao valor da compra?

**Coluna nova criada retornando a porcentagem do frete**

- Qual a porcentagem dos pedidos realizados com Voucher?

**Respondido por meio do gráfico de pizza**

```python
#Quantidade de brinquedos comprados no valor abaixo de 100,00
df_olist.loc[(df_olist['product_category_name'] == 'brinquedos') & (df_olist['price'] <= 100.00)]
```

```python
#Quantidade de birnquedos comprados no valor acima de 100,00
df_olist.loc[(df_olist['product_category_name'] == 'brinquedos') & (df_olist['price'] > 100.00)]
```

```python
#Qual a porcentagem do frete em relação ao valor da compra?
df_olist['Porcentagem'] = (df_olist['freight_value'] / df_olist['price']) * 100
df_olist
```

```python
#Criar uma nova coluna baseada na coluna "price" para Classificar os preços
df_olist.loc[df_olist['price'] < 100, 'Classificaçao_preços'] ='Econômico'
df_olist.loc[(df_olist['price'] > 100) & (df_olist['price']  <= 300),'Classificaçao_preços'] = 'Razoável'
df_olist.loc[df_olist['price'] > 300,'Classificaçao_preços'] = 'Alto Custo'
df_olist
```

```python
#Criar uma nova coluna baseada na coluna "payment_type" para Classificar os Tipos de pagamentos
df_olist.loc[df_olist['payment_type']=='credit_card', 'Tipo_de_pagamento']= 'Cartao de crédito'
df_olist.loc[df_olist['payment_type']=='boleto', 'Tipo_de_pagamento']= 'Boleto'
df_olist.loc[df_olist['payment_type']=='voucher', 'Tipo_de_pagamento']= 'Voucher'
df_olist.loc[df_olist['payment_type']=='debit_card', 'Tipo_de_pagamento']= 'Cartao de Débito'
df_olist
```

```python
#visualização que retorna Tipo de pagamento por classificaçao de preço
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(data = df_olist,
              x = 'Classificaçao_preços',
            hue= 'Tipo_de_pagamento')
plt.title('Tipo de pagamento por classificaçao de preços')
plt.xlabel('Classificaçao_preços')
plt.ylabel('Contagem')
plt.show()
```

```python
##visualização que retorna a porcentagem dos tipo de pagamentos 
#kind='pie' especifica que o tipo de gráfico é um gráfico de pizza.
#autopct='%1.1f%%' adiciona rótulos percentuais a cada fatia da pizza, com uma precisão de uma casa decimal.
df_olist['Tipo_de_pagamento'].value_counts().plot(
    kind='pie', 
    autopct='%1.1f%%')
plt.title('Porcentagem de Cada Tipo de Pagamento')
plt.show()
```
