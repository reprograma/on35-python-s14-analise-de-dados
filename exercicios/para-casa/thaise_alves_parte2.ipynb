## Dados Olist

```python
import pandas as pd
```

#### Tipos de pagamentos

```python
df_pagamentos = pd.read_csv('dados/olist_order_payments_dataset.csv')
df_pagamentos
```

#### Pedidos

```python
df_pedidos = pd.read_csv('dados/olist_orders_dataset.csv')
df_pedidos
```

##### Tipos de pagamento + Pedidos

```python
df_olist_pagamento_pedidos = pd.merge(df_pagamentos, df_pedidos, on=['order_id'], how='inner')
df_olist_pagamento_pedidos
```

#### Itens

```python
df_itens_pedido = pd.read_csv('dados/olist_order_items_dataset.csv')
df_itens_pedido
```

##### Tipos de pagamento + Pedidos + Itens

```python
df_olist_itens_pagos = pd.merge(df_olist_pagamento_pedidos, df_itens_pedido, on=['order_id'], how='inner')
df_olist_itens_pagos
```

#### Produtos

```python
df_produtos = pd.read_csv('dados/olist_products_dataset.csv')
df_produtos
```

##### Tipos de pagamento + Pedidos + Itens + Produtos

```python
df_olist_pagamento_produtos = pd.merge(df_olist_itens_pagos, df_produtos, on=['product_id'], how='inner')
df_olist_pagamento_produtos
```

### Perguntas

1. Qual o mÃ©todo de pagamento mais utilizado pelos clientes? A escolha desse mÃ©todo Ã© influÃªnciada pelo valor do produto?

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
palette = {'credit_card': 'purple', 'boleto': 'green', 'debit_card': 'blue', 'voucher': 'orange'}

plt.figure(figsize=(8, 6))
sns.countplot(data=df_olist_pagamento_produtos, x='payment_type', palette=palette)
plt.title('MÃ©todos de pagamento utilizados pelos clientes')
plt.xlabel('MÃ©todo de pagamento')
plt.ylabel('NÃºmero de pedidos')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df_olist_pagamento_produtos, x='payment_type', y='price', palette='tab10')
plt.title('RelaÃ§Ã£o entre preÃ§o do produto e tipo de pagamento')
plt.xlabel('Tipo de pagamento')
plt.ylabel('PreÃ§os dos produtos')

plt.show()
```

O cartÃ£o de crÃ©dito Ã© o mÃ©todo de pagamento mais usado nas compras realizadas e Ã© observada sua preferÃªncia entre os clientes em usÃ¡-lo com valores maiores.
Um ponto interessante Ã© que embora o cartÃ£o de dÃ©bito ser o mÃ©todo de pagamento menos utilizado Ã© visto que ele Ã© bastante usado nas compras com os segundo maiores valores.

2. O valor pago pelo produto influencia no nÃºmero de parcelas utilizadas na hora da compra?

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=df_olist_pagamento_produtos, x='payment_installments', y='payment_value', palette='tab10')
plt.title('RelaÃ§Ã£o entre preÃ§o pago para o produto e nÃºmero de parcelas feitas na compra')
plt.xlabel('Parcelas')
plt.ylabel('PreÃ§os')

plt.show()
```

Ã‰ observado que o maior nÃºmero de parcelas (24 vezes) sÃ£o preferÃªncia no pagamento de produtos mais caros.
