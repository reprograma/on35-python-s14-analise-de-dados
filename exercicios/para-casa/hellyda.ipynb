```python
import pandas as pd
```

```python
df_pedidos = pd.read_csv('dados/olist_orders_dataset.csv')
df_pedidos
```

```python
df_itens_pedido = pd.read_csv('dados/olist_order_items_dataset.csv')
df_itens_pedido
```

```python
df_olist_pedidos = pd.merge(df_pedidos, df_itens_pedido, on=['order_id'], how='inner')
df_olist_pedidos
```

```python
df_consumidor = pd.read_csv('dados/olist_customers_dataset.csv')
df_consumidor
```

```python
df_olist_pedidos_uf = pd.merge(df_olist_pedidos, df_consumidor, on=['customer_id'], how='inner')
df_olist_pedidos_uf
```

```python
df_produto = pd.read_csv('dados/olist_products_dataset.csv')
df_produto
```

```python
df_olist_pedidos_produtos = pd.merge(df_olist_pedidos_uf, df_produto, on=['product_id'], how='inner')

df_olist_pedidos_produtos
```

```python
df_olist_pedidos_produtos.info()
```

```python
df_reduzido  = df_olist_pedidos_produtos[['order_id', 'order_status', 'price', 'freight_value','customer_state','product_category_name']]
df_reduzido
```

```python
df_reduzido.info()
```

```python
df_reduzido['order_status'].unique()
```

```python
df_reduzido = df_reduzido.loc[df_reduzido['order_status'] == 'delivered']

# Mostra o DataFrame 'df_pedidos' na tela, agora com apenas os pedidos entregues.
# 'df_pedidos' Ã© o nome do DataFrame que queremos mostrar.
df_reduzido
```

```python
df_reduzido.info()
```

```python
df_reduzido.to_csv('base_final_s14_olist.csv', index=False)
```

```python
df_estados = df_reduzido.groupby('customer_state').agg({'order_id': 'nunique'})
df_estados.sort_values(by='order_id', ascending=False)
```

```python
df_categorias = df_reduzido.groupby('product_category_name').agg({'order_id': 'nunique'})
df_categorias.sort_values(by='order_id', ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# definir tamanho da figura
plt.figure(figsize=(12, 8))

# instanciar o tipo de plotagem, no caso, um countplot usando x como as faixas etÃ¡rias
sns.countplot(data = df_reduzido,
              x = 'customer_state')

# definir texto do tÃ­tulo e eixos x e y
plt.title('Consumo por Estado')
plt.xlabel('Estados')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# Selecionar as 15 categorias mais frequentes
top_15_categorias = df_reduzido['product_category_name'].value_counts().nlargest(15).index

# Filtrar o DataFrame para conter apenas essas categorias
df_top_15 = df_reduzido[df_reduzido['product_category_name'].isin(top_15_categorias)]

sns.countplot(data = df_top_15, x = 'product_category_name')

plt.title('Principais categorias de Consumo')
plt.xlabel('Categorias')
plt.ylabel('Contagem')

plt.xticks(rotation=45) 
plt.show()
```
